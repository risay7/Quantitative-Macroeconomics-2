```python
import numpy as np
import matplotlib.pyplot as plt

# 
sigma = 1.5
beta = 0.98
rho = 0.6
sigma_e = 0.6
r = 0.04
w = 1

# 
h_L = np.exp(-sigma_e)
h_H = np.exp(sigma_e)

# 
a_t_grid = np.linspace(0, 20, 200)

# 
def calc_saving_rate(a_t, h):
    c_t = w * h + (1 + r) * a_t - a_t / (1 + beta)
    a_t_plus_1 = (beta * (w * h + (1 + r) * a_t - c_t))
    saving_rate = a_t_plus_1 / (c_t + a_t_plus_1)
    return saving_rate

saving_rate_L = [calc_saving_rate(a, h_L) for a in a_t_grid]
saving_rate_H = [calc_saving_rate(a, h_H) for a in a_t_grid]

# 
plt.plot(a_t_grid, saving_rate_L, label='Low Productivity (h_L)')
plt.plot(a_t_grid, saving_rate_H, label='High Productivity (h_H)')
plt.xlabel('Current Assets (a_t)')
plt.ylabel('Saving Rate (a_{t+1} / (c_t + a_{t+1}))')
plt.legend()
plt.title('Saving Rate vs Current Assets')
plt.grid(True) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
sigma = 1.5
beta = 0.98
rho = 0.6
sigma_e = 0.6
r = 0.04
w = 1
tau_k = 0.3  # 

# 
h_L = np.exp(-sigma_e)
h_H = np.exp(sigma_e)

# 
a_t_grid = np.linspace(0, 20, 200)

# 
def calc_saving_rate_with_tax(a_t, h):
    c_t = w * h + (1 + (1 - tau_k) * r) * a_t - a_t / (1 + beta)
    a_t_plus_1 = (beta * (w * h + (1 + (1 - tau_k) * r) * a_t - c_t))
    saving_rate = a_t_plus_1 / (c_t + a_t_plus_1)
    return saving_rate

saving_rate_L_tax = [calc_saving_rate_with_tax(a, h_L) for a in a_t_grid]
saving_rate_H_tax = [calc_saving_rate_with_tax(a, h_H) for a in a_t_grid]

# 
def calc_saving_rate(a_t, h):
    c_t = w * h + (1 + r) * a_t - a_t / (1 + beta)
    a_t_plus_1 = (beta * (w * h + (1 + r) * a_t - c_t))
    saving_rate = a_t_plus_1 / (c_t + a_t_plus_1)
    return saving_rate

saving_rate_L = [calc_saving_rate(a, h_L) for a in a_t_grid]
saving_rate_H = [calc_saving_rate(a, h_H) for a in a_t_grid]

# 
plt.plot(a_t_grid, saving_rate_L, label='Low Productivity (h_L) - No Tax', linestyle='--')
plt.plot(a_t_grid, saving_rate_H, label='High Productivity (h_H) - No Tax', linestyle='--')
plt.plot(a_t_grid, saving_rate_L_tax, label='Low Productivity (h_L) - With Tax')
plt.plot(a_t_grid, saving_rate_H_tax, label='High Productivity (h_H) - With Tax')
plt.xlabel('Current Assets (a_t)')
plt.ylabel('Saving Rate (a_{t+1} / (c_t + a_{t+1}))')
plt.legend()
plt.title('Saving Rate vs Current Assets (With and Without Tax)')
plt.grid(True)  # 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
sigma = 1.5
beta_low = 0.1  # 
beta_high = 0.98
rho = 0.6
sigma_e = 0.6
r = 0.04
w = 1

# 
h_L = np.exp(-sigma_e)
h_H = np.exp(sigma_e)

# 
a_t_grid = np.linspace(0, 20, 200)

# 
def calc_saving_rate_low_beta(a_t, h):
    c_t = w * h + (1 + r) * a_t - a_t / (1 + beta_low)
    a_t_plus_1 = (beta_low * (w * h + (1 + r) * a_t - c_t))
    saving_rate = a_t_plus_1 / (c_t + a_t_plus_1)
    return saving_rate

saving_rate_L_low_beta = [calc_saving_rate_low_beta(a, h_L) for a in a_t_grid]
saving_rate_H_low_beta = [calc_saving_rate_low_beta(a, h_H) for a in a_t_grid]

# 
def calc_saving_rate(a_t, h):
    c_t = w * h + (1 + r) * a_t - a_t / (1 + beta_high)
    a_t_plus_1 = (beta_high * (w * h + (1 + r) * a_t - c_t))
    saving_rate = a_t_plus_1 / (c_t + a_t_plus_1)
    return saving_rate

saving_rate_L = [calc_saving_rate(a, h_L) for a in a_t_grid]
saving_rate_H = [calc_saving_rate(a, h_H) for a in a_t_grid]

# 
plt.plot(a_t_grid, saving_rate_L, label='Low Productivity (h_L) - High Beta', linestyle='--')
plt.plot(a_t_grid, saving_rate_H, label='High Productivity (h_H) - High Beta', linestyle='--')
plt.plot(a_t_grid, saving_rate_L_low_beta, label='Low Productivity (h_L) - Low Beta')
plt.plot(a_t_grid, saving_rate_H_low_beta, label='High Productivity (h_H) - Low Beta')
plt.xlabel('Current Assets (a_t)')
plt.ylabel('Saving Rate (a_{t+1} / (c_t + a_{t+1}))')
plt.legend()
plt.title('Saving Rate vs Current Assets (High Beta vs Low Beta)')
plt.grid(True)  # 
plt.show()
```
